NetworkX provides:

```python
cities = {'MPLS': ['MADISON','DES_MOINES','CHICAGO'],'CHICAGO':['DETROIT','NASHVILLE','MPLS'],
          'MADISON':['MPLS'],'DETROIT':['CHICAGO','DAYTON','DES_MOINES'],'CLOMBUS':['DAYTON'],
          'DAYTON': ['CLOMBUS','DETROIT'],'DES_MOINES': ['MPLS','CHICAGO'],'NASHVILLE': ['CHICAGO']}
```

Creating the graph of the cities with labels on them

--------------------------------------------------------------------------------------------------------------

NETWORK OF FIVE CITIES INTERSTATE

```python
import networkx as nx
G = nx.Graph(cities)
nx.draw(G, with_labels=True)
```

```python
nx.adjacency_matrix(G).todense()
```

```python
d = dict(G.degree())
```

```python
import matplotlib.pylab as plt
from matplotlib.pyplot import figure

figure(figsize=(10, 8), dpi=80)

lists = sorted(d.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.bar(x, y, color = 'maroon')
plt.show()
```
